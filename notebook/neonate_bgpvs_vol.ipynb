```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from skimage.filters import frangi, hessian
import skimage.morphology as mp
import os
```

### Load T2 image

```python
t2 = nib.load('../data/PVAS_Data/Nifti/T2_20240405092639_6.nii')
t2img = t2.get_fdata()
```

### Load Tissue Segmentation and Extract BG ROI

```python
seg = nib.load('../data/PVAS_Data/WM_Masks/ip2T2_20240405092639_6.nii') # infant freesurfer segmentation
wm = seg.get_fdata()
'''
bg = np.zeros_like(seg)
bg[seg==40] = 1
bg[seg==41] = 1
bg[seg==46] = 1
bg[seg==47] = 1
bg = mp.binary_closing(bg, mp.ball(3))
'''
```

```python
sn = 200
plt.imshow(np.rot90(t2img[72:-72,96:-92,sn]),cmap='gray')
plt.axis('off')
plt.title('T2w MRI')
plt.show()
plt.imshow(np.rot90(wm[72:-72,96:-64,sn]),cmap='tab20',interpolation='nearest')
plt.axis('off')
plt.title('White Matter Segmentation')
plt.show()
```

### Apply Frangi Filtering to T2

```python
t2img_f1 = frangi(t2img, sigmas=(1, 2), scale_step=1,black_ridges=False)
```

```python
plt.imshow(np.rot90(t2img_f1[:,:,180]),cmap='gray',vmax=0.1)
plt.axis('off')
plt.title('Frangi-Filtered T2w MRI')
plt.show()
```

```python
t2img_f2 = np.copy(t2img_f1)
t2img_f2[wm==0]=0
plt.imshow(np.rot90(t2img_f2[:,:,180]),cmap='gray',vmax=0.1)
plt.axis('off')
plt.title('Frangi-Filtered T2w MRI')
plt.show()
```

### Thresholding using WM ROI mean and standard deviation values

```python
wm_mean = np.mean(t2img_f2[wm==1])
wm_std = np.std(t2img_f2[wm ==1])
wmpvs = t2img_f2>(wm_mean+1*wm_std)
wmpvs = mp.binary_opening(wmpvs,mp.ball(1))
```

```python
plt.imshow(np.rot90(t2img[72:-72,96:-64,170]),cmap='gray',vmax=32)
plt.imshow(np.rot90(wmpvs[72:-72,96:-64,170]),cmap='hot',vmax=2,alpha=0.5)
plt.axis('off')
plt.title('WM PVS mask')
plt.show()
```

### Calculate BG PVS volume

```python
plt.imshow(np.rot90(np.sum(wmpvs[:,:,:],axis=2)),cmap='hot',vmax=16)
plt.axis('off')
plt.title('Max. Intensity Projection of WM PVS mask')
plt.show()
wmpvs_vol = np.prod(t2.header['pixdim'][1:4]) * len(wmpvs[wmpvs>0])
print(f'WM PVS volume: {wmpvs_vol} mm3')
```
